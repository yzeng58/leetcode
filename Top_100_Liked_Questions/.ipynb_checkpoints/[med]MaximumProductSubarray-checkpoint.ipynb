Given an integer array nums, find the contiguous subarray within an array (containing at least one number) which has the largest product.

```python
# time limit exceeded

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        max_prod = -float('inf')
        prod_array = []
        for i in range(len(nums)-1):
            prod_array.append([nums[i]])
            max_prod = max(nums[i], max_prod)
            for j in range(i+2, len(nums)+1):
                prod_array[i].append(prod_array[i][j-i-2] * nums[j-1])
                max_prod = max(prod_array[i][-1], max_prod)
        return max(max_prod, nums[-1])
```

```python
# my solution
# Runtime: 48 ms, faster than 93.50% of Python3 online submissions for Maximum Product Subarray.
# Memory Usage: 14.7 MB, less than 20.34% of Python3 online submissions for Maximum Product Subarray.

from functools import reduce

class Solution:
    def prod(self, nums):
        while -float('inf') in nums:
            nums.remove(-float('inf'))
        if len(nums) == 0:
            return -float('inf')
        return reduce(lambda x,y: x*y, nums)
    
    def maxProduct(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return -float('inf')
        
        all_prod = self.prod(nums)
        max_prod = all_prod
        
        if all_prod == 0:
            zero_idx = nums.index(0)
            return max(0, self.maxProduct(nums[:zero_idx]), self.maxProduct(nums[zero_idx+1:]))  
        elif all_prod > 0:
            return all_prod
        
        interval_prod = [-float('inf')]
        pos_idx = []
            
        for i in range(len(nums)):
            if nums[i] < 0:
                pos_idx.append(nums[i])
                interval_prod.append(-float('inf'))
            else:
                if interval_prod[-1] == -float('inf'):
                    interval_prod[-1] = 1
                interval_prod[-1] *= nums[i]
        
        for i in range(len(pos_idx)):
            max_prod = max(max_prod, self.prod(interval_prod[:i+1] + pos_idx[:i]), 
                           self.prod(interval_prod[i+1:] + pos_idx[i+1:]))
        
        return max_prod
```

```python
# official solution: dynamic programming
# Runtime: 52 ms, faster than 80.93% of Python3 online submissions for Maximum Product Subarray.
# Memory Usage: 14.3 MB, less than 79.93% of Python3 online submissions for Maximum Product Subarray.

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0

        max_so_far = nums[0]
        min_so_far = nums[0]
        result = max_so_far

        for i in range(1, len(nums)):
            curr = nums[i]
            temp_max = max(curr, max_so_far * curr, min_so_far * curr)
            min_so_far = min(curr, max_so_far * curr, min_so_far * curr)

            max_so_far = temp_max

            result = max(max_so_far, result)

        return result
```
