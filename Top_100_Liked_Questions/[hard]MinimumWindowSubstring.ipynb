Given two strings s and t, return the minimum window in s which will contain all the characters in t. If there is no such window in s that covers all characters in t, return the empty string "".

Note that If there is such a window, it is guaranteed that there will always be only one unique minimum window in s.

```python
# time limit exceed

class Solution:
    def wordCount(self, s: str) -> dict:
        d = {}
        for c in s:
            if not c in d:
                d[c] = 0
            d[c] += 1
        return d
    
    def isContain(self, s1: str, s2: str) -> bool:
        d1 = self.wordCount(s1)
        d2 = self.wordCount(s2)
        for c in d2:
            if not c in d1 or d1[c] < d2[c]:
                return False
        return True
    
    def intersection(self, s2: str, s1: str) -> list:
        d1 = self.wordCount(s1)
        d2 = self.wordCount(s2)
        d = []
        for c in d2:
            if c in d1:
                d.extend([c] * min(d2[c], d1[c]))
        return d
    
    def minWindow(self, s: str, t: str) -> str:
        if not self.isContain(s, t):
            return ""
        min_window_idx = [0,len(s)]
        
        d = [[list()] * len(s)] * len(s)

        for i in reversed(range(len(s))):
            for j in range(i+1, len(s)+1):
                if i == 0:
                    d[i][j-1] = self.intersection(t, s[i:j])
                else:
                    if self.isContain(t, ''.join(d[i][j-2] + [s[j-1]])):
                        d[i][j-1] = d[i][j-2] + [s[j-1]]
                    else:
                        d[i][j-1] = d[i][j-2]

                if sorted(d[i][j-1]) == sorted(t) and min_window_idx[1] - min_window_idx[0] > j-i:
                    min_window_idx = [i,j]

        if min_window_idx[1] == len(s):
            return s[min_window_idx[0]:]
        return s[min_window_idx[0]: min_window_idx[1]]                         
```

```python
# time limite exceed

from copy import deepcopy
from functools import reduce

class Solution:
    def isContain(self, s1: str, s2: str) -> bool:
        d1 = self.wordCount(s1)
        d2 = self.wordCount(s2)
        for c in d2:
            if not c in d1 or d1[c] < d2[c]:
                return False
        return True
    
    def wordCount(self, s: str) -> dict:
        d = {}
        for c in s:
            if not c in d:
                d[c] = 0
            d[c] += 1
        return d
    
    def wordIndex(self, s: str, word: list) -> dict:
        d = {}
        for i in range(len(s)):
            if s[i] in word:
                if not s[i] in d:
                    d[s[i]] = []
                d[s[i]].append(i)
        return d
    
    def minWindow(self, s: str, t: str) -> str:
        if not self.isContain(s, t):
            return ""
        
        word_count_t = self.wordCount(t)
        word_idx_s = self.wordIndex(s, list(word_count_t.keys()))
        
        min_window = [0, len(s)]
        
        idxs = sorted(reduce(lambda x, y: x+y, word_idx_s.values()))
        
        for i in range(len(s)):
            word_count_t_i = deepcopy(word_count_t)
            for idx in idxs:
                if idx < i or not s[idx] in word_count_t_i:
                    continue
                word_count_t_i[s[idx]] -= 1
                if word_count_t_i[s[idx]] == 0:
                    del word_count_t_i[s[idx]]
                if len(word_count_t_i) == 0:
                    if idx - i < min_window[1] - min_window[0] - 1:
                        min_window = [i, idx + 1]
        return s[min_window[0]: min_window[1]]
```

```python
# official solution
# Runtime: 132 ms, faster than 37.19% of Python3 online submissions for Minimum Window Substring.
# Memory Usage: 25.3 MB, less than 5.08% of Python3 online submissions for Minimum Window Substring.

from collections import Counter

class Solution:
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        if not t or not s:
            return ""

        dict_t = Counter(t)

        required = len(dict_t)

        # Filter all the characters from s into a new list along with their index.
        # The filtering criteria is that the character should be present in t.
        filtered_s = []
        for i, char in enumerate(s):
            if char in dict_t:
                filtered_s.append((i, char))

        l, r = 0, 0
        formed = 0
        window_counts = {}

        ans = float("inf"), None, None

        # Look for the characters only in the filtered list instead of entire s. This helps to reduce our search.
        # Hence, we follow the sliding window approach on as small list.
        while r < len(filtered_s):
            character = filtered_s[r][1]
            window_counts[character] = window_counts.get(character, 0) + 1

            if window_counts[character] == dict_t[character]:
                formed += 1

            # If the current window has all the characters in desired frequencies i.e. t is present in the window
            while l <= r and formed == required:
                character = filtered_s[l][1]

                # Save the smallest window until now.
                end = filtered_s[r][0]
                start = filtered_s[l][0]
                if end - start + 1 < ans[0]:
                    ans = (end - start + 1, start, end)

                window_counts[character] -= 1
                if window_counts[character] < dict_t[character]:
                    formed -= 1
                l += 1    

            r += 1    
        return "" if ans[0] == float("inf") else s[ans[1] : ans[2] + 1]
```
