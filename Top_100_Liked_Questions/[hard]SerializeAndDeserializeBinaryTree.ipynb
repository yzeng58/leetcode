Serialization is the process of converting a data structure or object into a sequence of bits so that it can be stored in a file or memory buffer, or transmitted across a network connection link to be reconstructed later in the same or another computer environment.

Design an algorithm to serialize and deserialize a binary tree. There is no restriction on how your serialization/deserialization algorithm should work. You just need to ensure that a binary tree can be serialized to a string and this string can be deserialized to the original tree structure.

Clarification: The input/output format is the same as how LeetCode serializes a binary tree. You do not necessarily need to follow this format, so please be creative and come up with different approaches yourself.

```python
# my solution
# Runtime: 1128 ms, faster than 5.07% of Python3 online submissions for Serialize and Deserialize Binary Tree.
# Memory Usage: 19.1 MB, less than 27.35% of Python3 online submissions for Serialize and Deserialize Binary Tree.

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        queue = deque([root])
        output = []
        
        while queue and set(queue) != {None}:
            current = queue[0]
            if current == None:
                output.append("")
                queue.popleft()
            else:
                queue.append(current.left)
                queue.append(current.right)
                output.append(str(queue.popleft().val))
                
        return ' '.join(output)
            

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        
        nums = data.split(' ')
        
        root = TreeNode(int(nums[0])) if not nums[0] in ['None', ''] else None
        queue = [root]
        children = 0
        
        for i in range(1, len(nums)):
            if nums[i] == '':
                nums[i] = None
            else:
                nums[i] = int(nums[i])
                
            if children == 2:
                queue.pop(0)
                children = 0
            
            if nums[i] == None:
                children += 1
                continue
                
            queue.append(TreeNode(nums[i]))
            if children == 0:
                queue[0].left = queue[-1]
            else:
                queue[0].right = queue[-1]
                
            children += 1
                
        return root    
```

```python
# Your Codec object will be instantiated and called as such:
ser = Codec()
deser = Codec()
root = TreeNode.listToBinaryTree([1,2,3,None,None,4,5,6,7])
deser.deserialize(ser.serialize(root))
```
