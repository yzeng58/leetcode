```python
from typing import List
```

```python
class Solution:
    def generateHelper(self, left_par: int, n: int):
        gen_s = []
        if n == 0:
            return [")"*left_par]
        
        if left_par == 0:
            for s in self.generateHelper(left_par + 1, n-1):
                gen_s.append("(" + s)
        else:
            for s in self.generateHelper(left_par - 1, n):
                gen_s.append(")" + s)
            for s in self.generateHelper(left_par + 1, n-1):
                gen_s.append("(" + s)
                
        return gen_s
        
    def generateParenthesis(self, n: int) -> List[str]:
        left_par = 0
        return self.generateHelper(left_par, n)
```

```python
sol = Solution()
sol.generateParenthesis(8)
```
