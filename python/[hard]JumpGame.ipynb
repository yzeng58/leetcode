```python
# RecursionError, Time limit exceed

class Solution:
    def jumpRecursiveHelper(self, nums):
        if len(nums) <= 1:
            return 0
        min_step = 100
        for i in range(0, len(nums)-1)[::-1]:
            step = max(len(nums)-i-nums[i], 1)
            if step <= min_step:
                min_step = step
                jump_from = i
                
        return 1 + self.jumpRecursiveHelper(nums[:jump_from+1])
        
    def jump(self, nums: List[int]) -> int:
        return self.jumpRecursiveHelper(nums)
```
