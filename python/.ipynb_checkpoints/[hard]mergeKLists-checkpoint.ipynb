```python
# exceed time limit
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        prehead = ListNode(-10**5)
        prev = prehead 
        
        while True:
            min_idx = None
            remove_idx = []
            for i in range(len(lists)):
                if lists[i] == None:
                    remove_idx.append(i)
                else:
                    if min_idx == None or lists[min_idx].val > lists[i].val:
                        min_idx = i
               
            if min_idx == None:
                return prehead.next
            
            prev.next = lists[min_idx]
            lists[min_idx] = lists[min_idx].next
            prev = prev.next
            
            for i, idx in enumerate(remove_idx):
                lists.pop(idx-i)
                
        return prehead.next
```

```python
# use sort
class Solution:
    def removeNone(self, lists):
        remove_idx = []
        for i in range(len(lists))[::-1]:
            if lists[i] == None:
                remove_idx.append(i)
            else: 
                break
                
        for i in remove_idx:
            lists.pop(i)
            
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        prehead = ListNode(-10**5)
        prev = prehead 
        lists.sort(key = lambda x: x.val if x != None and x != [] else -10**5, reverse = True)
        self.removeNone(lists)
        
        while len(lists) > 0:
            prev.next = lists[-1]
            lists[-1] = lists[-1].next
            prev = prev.next
            
            if lists[-1] == None:
                lists.pop()
            else:
                lists.sort(key = lambda x: x.val, reverse = True)
                
        return prehead.next

```

```python
# better
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        amount = len(lists)
        interval = 1
        while interval < amount:
            for i in range(0, amount - interval, interval * 2):
                lists[i] = self.merge2Lists(lists[i], lists[i + interval])
            interval *= 2
        return lists[0] if amount > 0 else None

    def merge2Lists(self, l1, l2):
        head = point = ListNode(0)
        while l1 and l2:
            if l1.val <= l2.val:
                point.next = l1
                l1 = l1.next
            else:
                point.next = l2
                l2 = l1
                l1 = point.next.next
            point = point.next
        if not l1:
            point.next=l2
        else:
            point.next=l1
        return head.next
```
