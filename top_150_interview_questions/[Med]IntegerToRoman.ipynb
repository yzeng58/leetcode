# 12. Integer to Roman

Seven different symbols represent Roman numerals with the following values:


|Symbol|Value|
|---|---|
|I|1|
|V|5|
|X|10|
|L|50|
|C|100|
|D|500|
|M|1000|


Roman numerals are formed by appending the conversions of decimal place values from highest to lowest. Converting a decimal place value into a Roman numeral has the following rules:

If the value does not start with 4 or 9, select the symbol of the maximal value that can be subtracted from the input, append that symbol to the result, subtract its value, and convert the remainder to a Roman numeral.
If the value starts with 4 or 9 use the subtractive form representing one symbol subtracted from the following symbol, for example, 4 is 1 (I) less than 5 (V): IV and 9 is 1 (I) less than 10 (X): IX. Only the following subtractive forms are used: 4 (IV), 9 (IX), 40 (XL), 90 (XC), 400 (CD) and 900 (CM).
Only powers of 10 (I, X, C, M) can be appended consecutively at most 3 times to represent multiples of 10. You cannot append 5 (V), 50 (L), or 500 (D) multiple times. If you need to append a symbol 4 times use the subtractive form.
Given an integer, convert it to a Roman numeral.

 

Example 1:

> Input: num = 3749
>
> Output: "MMMDCCXLIX"
>
> Explanation:
>
> 3000 = MMM as 1000 (M) + 1000 (M) + 1000 (M)
>
> 700 = DCC as 500 (D) + 100 (C) + 100 (C)
>
> 40 = XL as 10 (X) less of 50 (L)
>
> 9 = IX as 1 (I) less of 10 (X)
>
> Note: 49 is not 1 (I) less of 50 (L) because the conversion is based on decimal places

Example 2:

> Input: num = 58
>
> Output: "LVIII"
>
> Explanation:
>
> 50 = L
>
> 8 = VIII

Example 3:

> Input: num = 1994
>
> Output: "MCMXCIV"
>
> Explanation:
>
> 1000 = M
>
> 900 = CM
>
> 90 = XC
>
> 4 = IV

```python
class Solution:
    def substract_dict(self, num: int):
        if num >= 100:
            return "C"
        elif num >= 10: 
            return "X"
        elif num >= 1: 
            return "I"
        
    def largest_quotient(self, num: int):
        if num == 0:
            return "", 0
        elif num // 1000 > 0:
            return "M", num - 1000
        elif num // 500 > 0:
            return "D", num - 500
        elif num // 100 > 0:
            return "C", num  - 100
        elif num // 50 > 0:
            return "L", num - 50
        elif num // 10 > 0:
            return "X", num - 10
        elif num // 5 > 0:
            return "V", num - 5
        else:
            return "I", num - 1
        
    def intToRoman_recursive(self, num: int, roman_format: str):
        first_digit = int(str(num)[0])
        
        if first_digit in [4, 9]:
            larger_num = int(str(first_digit+1) + str(num)[1:])
            new_digit, remaining = self.largest_quotient(larger_num)
            substract_prefix = self.substract_dict(num)
            roman_format += substract_prefix + new_digit
        else:
            new_digit, remaining = self.largest_quotient(num)
            roman_format += new_digit
        
        if remaining == 0:
            return roman_format
        else:
            return self.intToRoman_recursive(remaining, roman_format)
            
    def intToRoman(self, num: int) -> str:
        return self.intToRoman_recursive(num, "")
            
        
            
```

```python
class Solution:
    def intToRoman(self, num: int) -> str:
        # Creating Dictionary for Lookup
        num_map = {
            1: "I",
            5: "V",    4: "IV",
            10: "X",   9: "IX",
            50: "L",   40: "XL",
            100: "C",  90: "XC",
            500: "D",  400: "CD",
            1000: "M", 900: "CM",
        }
        
        # Result Variable
        r = ''
        
        
        for n in [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]:
            # If n in list then add the roman value to result variable
            while n <= num:
                r += num_map[n]
                num-=n
        return r
```
